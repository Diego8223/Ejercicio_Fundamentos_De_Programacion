# (25%) Ejercicio 2

La distancia entre dos puntos en un plano se calcula con la siguiente fórmula:

> $d(P_1, P_2)=\sqrt{{(x_2-x_1)}^2+{(y_2-y_1)}^2}$

Donde las coordenadas *x* y *y* indican la posición en la que se encuentra cada punto *P*.

Realiza un algoritmo que permita al usuario ingresar los cuatro valores de las coordenadas en orden (debes indicarle cuál coordenada está ingresando) y calcular la distancia *d*.

El resultado debe presentar dos cifras decimales y separador de miles.

Además, para el presente ejercicio necesitas importar la librería matemática ```math``` y usar sus componentes de potenciación y radicación.

````python
# Desarrolle su ejercicio aquí.
import math

# Solicitar coordenadas al usuario
x1 = float(input("Ingresa la coordenada x del punto P1: "))
y1 = float(input("Ingresa la coordenada y del punto P1: "))
x2 = float(input("Ingresa la coordenada x del punto P2: "))
y2 = float(input("Ingresa la coordenada y del punto P2: "))

# Calcular la distancia
d = math.sqrt(math.pow(x2 - x1, 2) + math.pow(y2 - y1, 2))

# Formatear el resultado
resultado = "{:,.2f}".format(d)

# Mostrar el resultado
print(f"La distancia entre los puntos P1({x1}, {y1}) y P2({x2}, {y2}) es: {resultado}")
````

# (25%) Ejercicio 4

Al final del bloque completaremos un programa que permita calcular la calificación final de un curso de acuerdo al porcentaje de ponderación y la calificación obtenida en cada tarea.

En este ejercicio debes usar [formularios](https://www.youtube.com/watch?v=vx7yOZ5_Cw8) si trabajas en Google Colab o la función ```input``` si trabajas en otro editor, para solicitar al usuario la calificación obtenida en cada una de las 4 tareas descritas en la tabla, y con esa información debes calcular la calificación final.


| Nombre actividad                        | Porcentaje  |
|-----------------------------------------|:-----------:|
| S25 - Actividad de aprendizaje Tarea 01 |     25%     |
| S25 - Actividad de aprendizaje Tarea 02 |     25%     |
| S25 - Actividad de aprendizaje Tarea 03 |     25%     |
| S25 - Actividad de aprendizaje Tarea 04 |     25%     |


````python
# Función para validar las calificaciones
def obtener_calificacion(tarea_numero):
    while True:
        try:
            calificacion = float(input(f"Ingrese la calificación obtenida en la Tarea {tarea_numero} (0-100): "))
            if 0 <= calificacion <= 100:
                return calificacion
            else:
                print("Error: La calificación debe estar entre 0 y 100. Intente de nuevo.")
        except ValueError:
            print("Error: Ingrese un número válido.")

# Solicitar calificaciones de las tareas
tarea_1 = obtener_calificacion(1)
tarea_2 = obtener_calificacion(2)
tarea_3 = obtener_calificacion(3)
tarea_4 = obtener_calificacion(4)

# Porcentaje de cada tarea
porcentaje_tarea = 100  # 25% para cada tarea

# Calcular la calificación final como promedio ponderado
calificacion_tarea_1 = tarea_1 * porcentaje_tarea
calificacion_tarea_2 = tarea_2 * porcentaje_tarea
calificacion_tarea_3 = tarea_3 * porcentaje_tarea
calificacion_tarea_4 = tarea_4 * porcentaje_tarea

# Sumar las calificaciones ponderadas para obtener la calificación final
calificacion_final = calificacion_tarea_1 + calificacion_tarea_2 + calificacion_tarea_3 + calificacion_tarea_4

# Mostrar resultados de cada tarea y la calificación final
print(f"\nResultados de las tareas:")
print(f"Tarea 01: {calificacion_tarea_1:.2f}")
print(f"Tarea 02: {calificacion_tarea_2:.2f}")
print(f"Tarea 03: {calificacion_tarea_3:.2f}")
print(f"Tarea 04: {calificacion_tarea_4:.2f}")
print(f"\nLa calificación final del curso es: {calificacion_final:.2f}")
````
